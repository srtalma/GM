

1.   Define Text Cleaning Function (clean_arabic)
2.   Load AraBERT Model and Tokenizer
3.  Define Embedding Generation Function (get_embedding)
4. Process Dataset in Chunks



```python
import pandas as pd
import re
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
import gc

# Function to clean Arabic text
def clean_arabic(text):
    text = re.sub(r'[\u064B-\u0652]', '', text)  # Remove tashkeel (diacritics)
    text = re.sub(r'[إأآا]', 'ا', text)  # Normalize different forms of Alef
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)  # Remove non-Arabic characters
    return text

# Load AraBERT Model and Tokenizer
model_name = "aubmindlab/bert-base-arabertv2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Function to generate embeddings for text
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].cpu().numpy()

# Process data in smaller chunks =366 chunk all
chunk_size = 5000

# Read the dataset in small chunks
for chunk_idx, chunk in enumerate(pd.read_csv('/content/Arabic Poem Comprehensive Dataset (APCD).csv', chunksize=chunk_size)):
    # Clean the text in this chunk
    chunk['clean_verse'] = chunk['البيت'].apply(clean_arabic)

    # Generate embeddings for this chunk
    chunk['verse_embedding'] = chunk['clean_verse'].apply(get_embedding)

    # Save the embeddings for this chunk to a .npy file
    chunk_embeddings = np.vstack(chunk['verse_embedding'].values)
    np.save(f'chunk_embeddings_{chunk_idx}.npy', chunk_embeddings)


    # Clear memory after processing the chunk
    del chunk_embeddings
    torch.cuda.empty_cache()
    gc.collect()

print("All chunks processed and saved!")
```

```python

metadata_columns = ['العصر', 'الشاعر', 'البحر']

# Top N poems to retrieve
top_n = 15
top_3 = 5

# Step 1: Retrieve the top N similar poems
print(f"Top {top_n} similar poems found:")
retrieved_poems = []
meter_count = {}

# Loop through the top N poems
for rank, idx in enumerate(top_indices[:top_n], start=1):
    poem = full_poem_df.iloc[idx]['البيت']  # Poem text
    similarity_score = similarities[idx]

    # Retrieve metadata
    metadata = {}
    for col in metadata_columns:
        if col in full_poem_df.columns:
            metadata[col] = full_poem_df.iloc[idx][col]
        else:
            metadata[col] = "Unknown"

    # Count the occurrence of each meter
    meter = metadata.get('البحر', 'Unknown')
    if meter != 'Unknown':
        meter_count[meter] = meter_count.get(meter, 0) + 1

    retrieved_poems.append((rank, idx, similarity_score, poem, metadata))

    print(f"{rank}. Poem Index: {idx}, Similarity Score: {similarity_score:.4f}")
    print(f"   Title: {metadata['العصر']}, Author: {metadata['الشاعر']}, Meter: {metadata['البحر']}")
    print(f"   Poem: {poem}\n")

# Step 2: Determine the most common meter among the top N poems
most_common_meter = max(meter_count, key=meter_count.get)
print(f"Most common meter: {most_common_meter}")

# Step 3: Filter poems by the most common meter and display the top 3
print(f"\nTop {top_3} poems with the meter: {most_common_meter}")
filtered_poems = [poem for poem in retrieved_poems if poem[4]['البحر'] == most_common_meter]

for rank, (original_rank, idx, similarity_score, poem, metadata) in enumerate(filtered_poems[:top_3], start=1):
    print(f"{rank}. Poem Index: {idx}, Similarity Score: {similarity_score:.4f}")
    print(f"   Title: {metadata['العصر']}, Author: {metadata['الشاعر']}, Meter: {metadata['البحر']}")
    print(f"   Poem: {poem}\n")
```

```python
# Assuming 'most_common_meter' and 'meter_rules' have been defined, as well as 'modelallam' for generation

# Retrieve top 5 poems that match the common meter and prepare retry limits
common_meter_rules = meter_rules.get(most_common_meter)
top_poems = [poem for poem in retrieved_poems if poem[4]['البحر'] == most_common_meter][:5]
max_retries = len(top_poems)

def generate_prompt(user_input, poem):
    """
    Generates the prompt using a provided poem as an example.
    """
    formatted_question = f"""
    <s> [INST]
    أنت نموذج لغوي ذكي، مهمتك هي كتابة بيت شعري جديد باستخدام الوزن: "{common_meter_rules['description']}" ({most_common_meter}).
    سيتم تقديم مثال من الشعراء باستخدام نفس الوزن. يجب أن تعبر عن موضوع أو مشاعر تتعلق بـ "{user_input}" وأن تتبع قواعد هذا البحر الشعري بدقة.

    ### مثال من قصيدة باستخدام {most_common_meter}:
    "{poem[3]}"
       الشاعر: {poem[4]['الشاعر']}, العصر: {poem[4]['العصر']}
       الوزن: {common_meter_rules['pattern']}

    ### قواعد البحر الشعري ({most_common_meter}):
    - التفعيلات: {common_meter_rules['pattern']}
    - التقطيع العروضي: {common_meter_rules['prosody']}
    - مثال: "{common_meter_rules['example']}"

    ### الآن، اكتب بيتًا شعريًا يعبر عن "{user_input}"، مع الالتزام بقواعد {most_common_meter}.
    [/INST]
    </s>
    """
    return formatted_question

def get_user_retry_input(attempt_number):
    """
    Prompt user to retry with a different example.
    """
    if attempt_number >= max_retries:
        print("لقد وصلت إلى الحد الأقصى للمحاولات.")
        return False

    user_response = input("هل ترغب في إعادة المحاولة باستخدام مثال مختلف؟ (نعم/لا): ").strip().lower()
    return user_response in ['نعم', 'yes']

def generate_poem(user_input):
    """
    Generates a poem based on user input with retry capability.
    """
    for attempt in range(max_retries):
        current_poem = top_poems[attempt]
        prompt = generate_prompt(user_input, current_poem)

        # Generate the response using the model
        generated_response = modelallam.generate_text(prompt=prompt, guardrails=False)

        # Display generated poem
        print(f"\nالقصيدة المولدة (المحاولة {attempt + 1}):\n{generated_response}\n")

        # Ask the user if they want to retry with a different example
        if attempt < max_retries - 1:
            retry = get_user_retry_input(attempt + 1)
            if retry:
                print("جاري إعادة المحاولة باستخدام مثال مختلف...\n")
                continue
            else:
                print("تم إنهاء العملية.")
                break
    else:
        print("لم يتم توليد قصيدة جديدة بعد المحاولات المتاحة.")

# Example usage
if __name__ == "__main__":لا
    generate_poem(user_input)
```

```python
# Function to provide Arabic prosody rules in Arabic
def provide_arudi_rules():
    # Arabic prosodic units and general rules
    prosody_rules = """
    ### القواعد الأساسية للتقطيع العروضي:

    - **الحركة**: أي صوت قصير (مثل "بَ") يُعبّر عنه بالرمز `/`.
    - **السكون**: الحرف الساكن (مثل "دْ") يُعبّر عنه بالرمز `–`.
    - **الحروف الطويلة (الألف والواو والياء الطويلة)**: الحروف الطويلة تُعبّر عنها بالرمز `–` (مثل "فَا").

    ### الوحدات العروضية:
    - **السبب الخفيف**: يتكون من حركة متبوعة بسكون، مثل "لَمْ" = `/ –`.
    - **السبب الثقيل**: يتكون من حركتين، مثل "عَلَى" = `/ /`.
    - **الوتد المجموع**: يتكون من حركتين متبوعتين بسكون، مثل "وَاحِدْ" = `/ – /`.
    - **الوتد المفروق**: يتكون من حركة، سكون، ثم حركة، مثل "مَوْلًى" = `/ – /`.

    ### أمثلة على التفعيلات:
    - **مُسْتَفْعِلُن**: `/ – / – /`.
    - **فَاعِلُن**: `/ – / –`.
    - **مَفَاعِيلُن**: `/ – – / –`.

    ### أمثلة على بحور الشعر:
    - **البحر الطويل**: فَعُولُن مَفَاعِيلُن فَعُولُن مَفَاعِيلُن.
    - **البسيط**: مُسْتَفْعِلُن فَاعِلُن مُسْتَفْعِلُن فَاعِلُن.
    """
    return prosody_rules
```

```python
# Function to retrieve top 3 poems by the poet's common meter
def top_poems_by_common_meter(df, poet_name):
    poet_df = df[df["الشاعر"] == poet_name]

    # Check if the poet has poems available
    if poet_df.empty:
        print("No poems available for this poet.")
        return None, None

    # Find the most common meter for this poet
    most_common_meter = poet_df["البحر"].mode()[0]

    # Filter poems to those with the common meter and get the top 3
    top_poems = poet_df[poet_df["البحر"] == most_common_meter].head(3)

    return top_poems[["الشاعر", "البحر", "البيت"]], most_common_meter
```

```python

common_meter_rules = meter_rules.get(most_common_meter)
top_poems = [poem for poem in retrieved_poems if poem[4]['البحر'] == most_common_meter][:5]
max_retries = len(top_poems)

def generate_prompt(user_input, poem):
    """
    Generates the prompt using a provided poem as an example.
    """
    formatted_question = f"""
    <s> [INST]
    أنت نموذج لغوي ذكي، مهمتك هي كتابة بيت شعري جديد باستخدام الوزن: "{common_meter_rules['description']}" ({most_common_meter}).
    سيتم تقديم مثال من الشعراء باستخدام نفس الوزن. يجب أن تعبر عن موضوع أو مشاعر تتعلق بـ "{user_input}" وأن تتبع قواعد هذا البحر الشعري بدقة.

    ### مثال من قصيدة باستخدام {most_common_meter}:
    "{poem[3]}"
       الشاعر: {poem[4]['الشاعر']}, العصر: {poem[4]['العصر']}
       الوزن: {common_meter_rules['pattern']}

    ### قواعد البحر الشعري ({most_common_meter}):
    - التفعيلات: {common_meter_rules['pattern']}
    - التقطيع العروضي: {common_meter_rules['prosody']}
    - مثال: "{common_meter_rules['example']}"

    ### الآن، اكتب بيتًا شعريًا يعبر عن "{user_input}"، مع الالتزام بقواعد {most_common_meter}.
    [/INST]
    </s>
    """
    return formatted_question

def get_user_retry_input(attempt_number):
    """
    Prompt user to retry with a different example.
    """
    if attempt_number >= max_retries:
        print("لقد وصلت إلى الحد الأقصى للمحاولات.")
        return False

    user_response = input("هل ترغب في إعادة المحاولة باستخدام مثال مختلف؟ (نعم/لا): ").strip().lower()
    return user_response in ['نعم', 'yes']

def generate_poem(user_input):
    """
    Generates a poem based on user input with retry capability.
    """
    for attempt in range(max_retries):
        current_poem = top_poems[attempt]
        prompt = generate_prompt(user_input, current_poem)

        # Generate the response using the model
        generated_response = modelallam.generate_text(prompt=prompt, guardrails=False)

        # Display generated poem
        print(f"\nالقصيدة المولدة (المحاولة {attempt + 1}):\n{generated_response}\n")

        # Ask the user if they want to retry with a different example
        if attempt < max_retries - 1:
            retry = get_user_retry_input(attempt + 1)
            if retry:
                print("جاري إعادة المحاولة باستخدام مثال مختلف...\n")
                continue
            else:
                print("تم إنهاء العملية.")
                break
    else:
        print("لم يتم توليد قصيدة جديدة بعد المحاولات المتاحة.")

# Example usage
if __name__ == "__main__":
    generate_poem(user_input)
```

```python
import pandas as pd
import re

def get_user_input():
    # تعريف الحقول التي سيتم إدخالها
    fields = {
        "البحر": {"type": str, "example": "الطويل"},
        "البيت": {"type": str, "example": "قفا نبك من ذكرى حبيب ومنزل"},
        "الشاعر": {"type": str, "example": "امرؤ القيس"},
        "اللهجة": {"type": str, "example": "فصحى"}
    }

    # نمط للتحقق من الأحرف العربية فقط
    arabic_pattern = re.compile(r'^[\u0600-\u06FF\s]+$')

    # جمع إدخالات المستخدم
    user_data = {}
    for field, specs in fields.items():
        while True:
            value = input(f"أدخل {field} ({specs['type'].__name__}, مثال: {specs['example']}): ").strip()

            # التحقق من عدم أن يكون الإدخال فارغاً
            if not value:
                print(f"إدخال {field} لا يمكن أن يكون فارغًا. حاول مرة أخرى.\n")
                continue

            # التحقق من أن الإدخال يحتوي فقط على أحرف عربية ومسافات
            if not arabic_pattern.match(value):
                print(f"إدخال {field} يجب أن يحتوي فقط على أحرف عربية ومسافات. حاول مرة أخرى.\n")
                continue

            # إذا كان الإدخال صحيحاً، يتم تخزينه
            user_data[field] = value
            print(f"تم إدخال {field} بنجاح: {value}\n")
            break

    return user_data

# إنشاء قائمة لتخزين جميع إدخالات المستخدم
data = []

# جمع إدخالات متعددة
while True:
    print("\nأدخل معلومات القصيدة (أو اكتب 'خروج' لإنهاء الإدخال):")

    # السماح للمستخدم بإنهاء الإدخال بكتابة 'خروج'
    user_choice = input("هل تريد إدخال بيانات جديدة؟ (نعم/خروج): ").strip().lower()
    if user_choice == 'خروج':
        break

    user_input = get_user_input()
    if user_input:
        data.append(user_input)

    # الاستفسار إذا كان المستخدم يرغب في إضافة إدخال آخر
    cont = input("هل ترغب في إضافة إدخال آخر؟ (نعم/لا): ").strip().lower()
    if cont == 'لا':
        break

# تحويل البيانات إلى إطار بيانات
df = pd.DataFrame(data)
print("\nالبيانات المدخلة:")
print(df)

# حفظ البيانات في ملف CSV مع ترميز يدعم العربية
df.to_csv('بيانات_القصائد_العربية.csv', index=False, encoding='utf-8-sig')
print("\nتم حفظ البيانات في ملف بيانات_القصائد_العربية.csv")
```
